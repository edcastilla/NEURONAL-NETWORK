```python
import numpy as np
import matplotlib.pyplot as plt
# import scipy as sc
```

```python
# FUNCIONES DE ACTIVACION

sigm = (lambda x: 1 / (1 + np.e ** (-x)),
        lambda x: x * (1 - x))

relu = lambda x: np.maximum(0, x)

_x = np.linspace(-5, 5, 100)
plt.figure()
plt.subplot(3,3,1)
plt.plot(_x, relu(_x))
plt.xlabel('rectificadora lineal (Relu)')
plt.subplot(3,3,2)
plt.plot(_x,sigm[0](_x))
plt.xlabel('funcion sigmoide')
plt.subplot(3,3,3)
plt.plot(_x,sigm[1](_x))
plt.xlabel('derivada funcion sigmoide')
```
